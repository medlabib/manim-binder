```python
"""
Cellular Process — Manim scene template.
Simplified cell diagram with labeled arrows for biology/biochemistry topics.
"""
from manim import *


class CellularProcess(Scene):
    def construct(self):
        self.camera.background_color = "#1a1a1a"

        title = Text("Cellular Process", font_size=36, color=WHITE).to_edge(UP, buff=0.4)
        self.play(Write(title))

        # --- Cell membrane (outer ellipse) ---
        membrane = Ellipse(width=6, height=4, color="#81c784", stroke_width=3, fill_opacity=0.08)
        membrane.move_to(ORIGIN)
        membrane_lbl = Text("Cell Membrane", font_size=18, color="#81c784")
        membrane_lbl.next_to(membrane, UP, buff=0.1)
        self.play(Create(membrane), Write(membrane_lbl))

        # --- Nucleus ---
        nucleus = Circle(radius=0.9, color="#4fc3f7", fill_opacity=0.3, stroke_width=2.5)
        nucleus.move_to([-0.5, 0.2, 0])
        nucleus_lbl = Text("Nucleus\n(DNA)", font_size=16, color="#4fc3f7").move_to(nucleus.get_center())
        self.play(GrowFromCenter(nucleus), Write(nucleus_lbl))

        # --- Mitochondrion ---
        mito = Ellipse(width=1.6, height=0.8, color="#ff8a65", fill_opacity=0.25, stroke_width=2)
        mito.move_to([1.8, -0.6, 0])
        mito_lbl = Text("Mitochondria", font_size=15, color="#ff8a65").next_to(mito, DOWN, buff=0.05)
        self.play(GrowFromCenter(mito), Write(mito_lbl))

        # --- Ribosome dots ---
        ribosomes = VGroup(*[
            Dot(radius=0.06, color="#fff176").move_to([x, y, 0])
            for x, y in [(0.9, 0.9), (1.4, 0.5), (1.0, 0.1), (2.0, 0.7)]
        ])
        ribo_lbl = Text("Ribosomes", font_size=15, color="#fff176").move_to([1.8, 1.2, 0])
        self.play(FadeIn(ribosomes), Write(ribo_lbl))

        # --- Process arrows ---
        arrow1 = Arrow(nucleus.get_right(), mito.get_left(), color=WHITE, buff=0.1, stroke_width=2)
        arrow1_lbl = Text("Energy\nSignal", font_size=14, color=WHITE).next_to(arrow1, UP, buff=0.05)
        self.play(GrowArrow(arrow1), Write(arrow1_lbl))

        # --- ATP output arrow leaving cell ---
        atp_arrow = Arrow([2.5, -0.6, 0], [3.8, -0.6, 0], color="#ff8a65", stroke_width=2.5)
        atp_lbl = Text("ATP →", font_size=18, color="#ff8a65").next_to(atp_arrow, UP, buff=0.05)
        self.play(GrowArrow(atp_arrow), Write(atp_lbl))

        self.wait(2)
```
